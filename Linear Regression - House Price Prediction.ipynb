**Features**

```python
X = boston_df.drop(['RAD','CHAS','MEDV'], axis = 1)
Y = boston_df['MEDV']
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
'''
The transformation is given by:

X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
X_scaled = X_std * (max - min) + min
where min, max = feature_range.
'''
```

### Cross Validation
* Split data into train & test randomly

```python
from sklearn.model_selection import train_test_split
# try random_state 0 to 9, and find best accuracy
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size = 0.2, random_state = 3)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression(normalize = True)
```

```python
# Fit() Training the  algorithm
lin_reg.fit(X_train, Y_train)
```

```python
lin_reg.score(X_test, Y_test)
```

Metrics of Regression

```python
yhat = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Variance of predicted value from actual value (%) (0-1)
r2_score(Y_test, yhat)
```

```python
mean_squared_error(Y_test, yhat)
```

```python
# RMSE
np.sqrt(mean_squared_error(Y_test, yhat))
```

```python
mean_absolute_error(Y_test, yhat)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Sample DataFrame with operational variables
data = {
    'ID': [1, 2, 3, 4, 5, 6],
    'batch': ['A', 'A', 'B', 'B', 'C', 'C'],
    'machine': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'Supplier': ['S1', 'S2', 'S1', 'S2', 'S1', 'S2'],
    'Results': [1, 0, 1, 0, 1, 1],
    'Op1': [10, 20, 10, 30, 20, 30],
    'Op2': [5, 3, 4, 2, 5, 2],
    'Op3': [7, 8, 5, 7, 8, 6],
    'Op4': [1, 3, 1, 3, 2, 2]
}

df = pd.DataFrame(data)

# Standardize the operational variables
scaler = StandardScaler()
op_vars = df[['Op1', 'Op2', 'Op3', 'Op4']]
op_vars_scaled = scaler.fit_transform(op_vars)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(op_vars_scaled)

# Add cluster labels to the DataFrame
df['Cluster'] = clusters

# Calculate the proportion of fails for each combination of batch, machine, supplier, and cluster
fail_proportion = df.groupby(['batch', 'machine', 'Supplier', 'Cluster'])['Results'].mean().reset_index()
fail_proportion.rename(columns={'Results': 'Fail_Proportion'}, inplace=True)

# Create the catplot
sns.relplot(x='batch', y='Fail_Proportion', hue='machine', row='Supplier', col='Cluster', data=fail_proportion, kind='bar', height=4, aspect=1.5)

# Display the plot
plt.show()
```

```python
# Create heatmaps for each cluster
for cluster in df['Cluster'].unique():
    cluster_data = df[df['Cluster'] == cluster]
    corr = cluster_data[['Op1', 'Op2', 'Op3', 'Op4']].corr()

    plt.figure(figsize=(8, 6))
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Pearson Correlation Coefficient Heatmap - Cluster {cluster}')
    plt.show()
```

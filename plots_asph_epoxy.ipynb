```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading Dataset
results_df = pd.read_csv('https://raw.githubusercontent.com/msht/aspha_epoxy/main/3results.csv?token=GHSAT0AAAAAABRWJ6W3AXGV53PXTT3LBGTMYQPFIGQ')
```

```python
results_df['name']
```

```python
results_df['name'][0:3] = "Pure\nepoxy"
```

```python
plt.figure(figsize = (2,1.8),dpi=600)
#sns.set(font_scale = 5)
sns.set_style("ticks", {"ytick.major.size":8})
sns.set_style({'font.family': 'Helvetica'})
#plt.style.use('nature')
my_pal = {"Pure\nepoxy": "#FF7F7F", "1wt%": "#997FFF", "3wt%":"#FFF34D", "5wt%":"#6EFFB9"}
ax = sns.boxplot(x=results_df['name'],y=results_df['E'],linewidth=0.3, palette=my_pal)
sns.stripplot(results_df['name'],results_df['E'],size=3, palette=my_pal)
#ax.grid(False)
ax.yaxis.set_tick_params(width=0.5,length=2, labelsize = 5)
ax.xaxis.set_tick_params(width=0.5,length=2, labelsize = 5)
ax.autoscale(tight=False)
ax.set(xlabel=None)
plt.tight_layout()
ax.set_ylabel("Young's Modulus (GPa)", fontsize = 5)
#plt.savefig('E.png')
```

```python
plt.figure(figsize = (2,1.8),dpi=600)
sns.set_style("ticks", {"ytick.major.size":8})
plt.style.use('nature')
my_pal = {"Pure\nepoxy": "#98f5ff", "1wt%": "#FFB482", "3wt%":"#54FF9F", "5wt%":"#FFFEA3"}
ax = sns.boxplot(x=results_df['name'],y=results_df['UTS'],linewidth=0.3, palette=my_pal)
sns.stripplot(results_df['name'],results_df['UTS'],size=3, palette=my_pal)
#ax.grid(False)
ax.yaxis.set_tick_params(width=0.5,length=2,labelsize=5)
ax.xaxis.set_tick_params(width=0.5,length=2,labelsize=5)
ax.autoscale(tight=False)
ax.set(xlabel=None)
sns.set(font="Helvetica")
plt.tight_layout()
ax.set_ylabel("Ultimate Tensile Strength (MPa)",fontsize=5)
plt.savefig('UTS.png')
```

```python
plt.figure(figsize = (1.8,1.8),dpi=600)
#sns.set_style('whitegrid')
plt.style.use('nature')
ax = sns.boxplot(results_df['name'],results_df['US'])
sns.stripplot(results_df['name'],results_df['US'])
ax.autoscale(tight=False)
ax.set(xlabel=None)
ax.set_ylabel("Strain at Failure (%)", fontsize = 7)
plt.savefig('US.png')
```

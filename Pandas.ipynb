```python
import pandas as pd
```

```python
# Array of data
val = [100, 200, 300, 400, 500, 600, 700]
```

###Series

```python
# In Python Series index positions starts from 0
# It is 1-D data types , default data type
pd.Series(data = val)
```

```python
# dtype = int8, int16, int32, int64
pd.Series(data = [100, 200, 300, 400, 500, 600, 700], index = ['a','b','c','d','e','f','g'], dtype = 'int16')
```

```python
# dtype = float16, float32, float64
pd.Series(data = [10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0], index = ['a','b','c','d','e','f','g'], dtype = 'int16')
```

```python
# Dictionary of Series Dataset
data_series = { 'Column1' : pd.Series(data = [100,200,300,400,500,600,700], index = ['a','b','c','e','f','g','h'], dtype = 'int16'),
               'Column2' : pd.Series(data = [20,40,60,80,100,120,140], index = ['a','b','c','e','f','g','i'], dtype = 'int16')}
```

```python
data_series
```

```python
# Type casted series, lists, arrays type of data into excel like datatype
df_1 = pd.DataFrame(data_series)
df_1.isnull()
```

```python
df_1.isnull().sum()
```

```python
# Dropping Rows , axis = 0
# To save changes we write inplace = True
movies_df.drop([0,1,2,3,4,5,6], axis = 0)
```

```python
# axis = 1 (For column operation)
movies_df.drop(['Film', 'Worldwide Gross'], axis = 1)
```

```python
# Boolean Condition
movies_df[movies_df['Year'] == 2008]
```

```python
# Or Condition
movies_df[(movies_df['Year'] == 2008) | (movies_df['Genre'] == 'Comedy')]
```

```python
# AND Condition
movies_df[(movies_df['Year'] == 2011) & (movies_df['Genre'] == 'Romance')]
```

```python
# Before ',' it slices rows & after ',' it slices columns using range
movies_df.loc[:, :]
```

```python
# Before ',' it slices rows & after ',' it slices columns using range
movies_df.loc[10:30 , : ] 
```

```python
# loc takes rows index as rows names & columns names to slice data
movies_df.loc[: , ['Film', 'Genre', 'Lead Studio', 'Audience score %', 'Profitability']]
```

```python
# loc takes rows index as rows names & columns names to slice data
movies_df.loc[10:20 , ['Film', 'Genre', 'Lead Studio', 'Audience score %', 'Profitability']]
```

```python
# iloc - i is for index positions
# It slices data using index positions
movies_df.iloc[10:20, ]
```

```python
movies_df.iloc[10:20, 2:7]
```

```python
movies_df.iloc[10:30:2, 2:7:2]
```

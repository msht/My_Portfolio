**Dictionary**

Dictonary are keys and values pair of declaration.

```python
dict_1 = {}
```

```python
# Dictionaries are keys and values pair of declaration
# Keys must be unique values
dict_1 = {'A':'Apple', 'B' : 'Ball', 'C' : 'Cat', 'D' : 'Dog'}
```

```python
fruit_basket = {'Apple': "Red", 'Orange': 'Orange', 'Grape' : 'Green', 'Blueberry' : 'Purple'}
```

```python
print(fruit_basket['Apple'])
```

```python
# Declare Multi-Dimentional Dictionary
fruits = { 'fruits_name' : {
    'Apple' : {
        'Color' : 'Red',
        'Shape' : 'Round'
    },
    'Orange' : {
        'Color' : 'Orange',
        'Shape' : 'Round'
    }
  }
}
```

```python
fruits['fruits_name']['Apple']
```
